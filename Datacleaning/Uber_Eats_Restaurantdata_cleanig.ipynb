**RATING NORMLAIZATION**

```python
import numpy as np

cleaned_rating = []

for x in data['rating_outof_5']:
  if x == 'NEW':
    cleaned_rating.append(np.nan)

  else:
    value = str(x).split('/')[0].strip()
    cleaned_rating.append(float(value))

data['rating_outof_5'] = cleaned_rating
```

```python
print(data['rating_outof_5'].unique())
```

**Feature engineering for rating categories**

```python
def rating_category(rating):
    if rating >= 4.0:
        return 'Excellent'
    elif rating >= 3.0:
        return 'Good'
    else:
        return 'Average'

data['rating_category'] = data['rating_outof_5'].apply(rating_category)
```

```python
data.rating_category
```

**Feature engineering for pricing segment**

```python
def pricing_segment(cost):

    if cost <= 500:
        return 'Budget'
    elif cost <= 1000:
        return 'Mid-Range'
    else:
        return 'Premium'

data['pricing_segment'] = data['approx_cost_for_2'].apply(pricing_segment)
```
